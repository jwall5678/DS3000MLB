```python
payroll_list = pd.read_html("http://www.stevetheump.com/Payrolls.htm", header = 0)


years = []
for i in range(2000, 2023):
    years.append(str(i))

years.reverse()
```

```python
payroll = {}

for i in range(len(years)):
    payroll[years[i]] = payroll_list[i+1].dropna(axis = 0)
    
```

**Cleaning Salaries**

```python
def uncomma(str_):
    string = str_.split(",")
    
    string = string[1][1:] + " " + string[0]
    
    return string
```

```python
for year in years:
    new_df = salaries[year]
    new_df["Name"] = new_df["Player"].apply(uncomma)
```
